```python
# Step 1: Mount Google Drive to access the dataset
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Load the dataset
import pandas as pd
file_path = '/content/drive/MyDrive/Price-Prediction---Project-4-/Resources/realtordataillinois.csv'
df = pd.read_csv(file_path)
df
```

```python
# Step 3: Filter the dataset for Illinois houses that are 'for_sale' or 'sold'
df_query = df.query("state=='Illinois' & (status=='for_sale' | status=='sold')")

df_query
```

```python
# Step 4: Drop rows with missing essential data
df_query.dropna(subset=['bed', 'bath', 'acre_lot', 'street', 'city', 'state', 'zip_code', 'house_size', 'prev_sold_date', 'brokered_by', 'status', 'price'], inplace=True)

df_query
```

```python
# Step 5: Create 'year' column from 'prev_sold_date'
df_query['year'] = df_query['prev_sold_date'].str[:4].astype(int)

df_query
```

```python
df_query["zip_code"].value_counts()
```

```python
df_dummies = pd.get_dummies(df_query, columns=['zip_code'], prefix='zip')
df_dummies
```

```python
X = df_dummies.drop(columns=['status','price','street','city',"state",'prev_sold_date','brokered_by'])
# price is what we want to predict
y = df_dummies['price']
y.head()
```

```python
# Use sklearn to split dataset
from sklearn.model_selection import train_test_split
import sklearn as skl
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 20

# define base model
def baseline_model():
# create model
 model = Sequential()
 model.add(Dense(hidden_nodes_layer1, input_shape=(number_input_features,), activation='relu'))
 model.add(Dense(hidden_nodes_layer2, activation='relu'))
 model.add(Dense(1))

 # Compile model
 model.compile(loss='mean_squared_error', optimizer='adam',metrics=["accuracy"])
 return model
# evaluate model
estimator = KerasRegressor(model=baseline_model, epochs=100, batch_size=32, verbose=0)
kfold = KFold(n_splits=5)

# Use R² score instead of accuracy
r2_scorer = make_scorer(r2_score)
results = cross_val_score(estimator, X_test_scaled,y_test, cv=kfold, scoring=r2_scorer)
print("Baseline: %.2f (%.2f) R2" % (results.mean(), results.std()))
```

```python
# Step 14: Fit the model on the training data
estimator.fit(X_train_scaled, y_train)

# Evaluate the model on the test data
test_loss = estimator.score(X_test_scaled, y_test)
print(f"Test loss (MSE): {test_loss}")
```

```python
from sklearn.metrics import r2_score

# Make predictions on the test set
y_pred = estimator.predict(X_test_scaled)

# Calculate R² score
r2 = r2_score(y_test, y_pred)
print(f"R² score: {r2}")
```
