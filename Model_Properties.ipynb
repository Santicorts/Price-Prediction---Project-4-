```python
# Mount Google Drive to access the dataset
from google.colab import drive
drive.mount('/content/drive')

# Load the dataset
import pandas as pd
file_path = '/content/drive/MyDrive/Price-Prediction---Project-4-/Resources/realtordataillinois.csv'
df = pd.read_csv(file_path)
df
```

```python
# Filter the dataset for Illinois houses that are 'for_sale' or 'sold'
df_query = df.query("state=='Illinois' & (status=='for_sale' | status=='sold')")

df_query
```

```python
# Drop rows with missing essential data
df_query.dropna(subset=['bed', 'bath', 'acre_lot', 'street', 'city', 'state', 'zip_code', 'house_size', 'prev_sold_date', 'brokered_by', 'status', 'price'], inplace=True)

df_query
```

```python
# Create 'year' column from 'prev_sold_date'
df_query['year'] = df_query['prev_sold_date'].str[:4].astype(int)

df_query
```

```python
df_query["zip_code"].value_counts()
```

```python
df_dummies = pd.get_dummies(df_query, columns=['zip_code'], prefix='zip')
df_dummies
```

```python
X = df_dummies.drop(columns=['status','price','street','city',"state",'prev_sold_date','brokered_by'])
# price is what we want to predict
y = df_dummies['price']
y.head()
```

```python
# Use sklearn to split dataset
from sklearn.model_selection import train_test_split
import sklearn as skl
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from scikeras.wrappers import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import make_scorer, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)

# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 20

# define base model
def baseline_model():
# create model
 model = Sequential()
 model.add(Dense(hidden_nodes_layer1, input_shape=(number_input_features,), activation='relu'))
 model.add(Dense(hidden_nodes_layer2, activation='relu'))
 model.add(Dense(1))

 # Compile model
 model.compile(loss='mean_squared_error', optimizer='adam',metrics=["accuracy"])
 return model
# evaluate model
estimator = KerasRegressor(model=baseline_model, epochs=100, batch_size=32, verbose=0)
kfold = KFold(n_splits=5)

# Use R² score instead of accuracy
r2_scorer = make_scorer(r2_score)
results = cross_val_score(estimator, X_test_scaled,y_test, cv=kfold, scoring=r2_scorer)
print("Baseline: %.2f (%.2f) R2" % (results.mean(), results.std()))
```

```python
# Fit the model on the training data
estimator.fit(X_train_scaled, y_train)

# Evaluate the model on the test data
test_loss = estimator.score(X_test_scaled, y_test)
print(f"Test (MSE): {test_loss}")
```

```python
from sklearn.metrics import r2_score

# Make predictions on the test set
y_pred = estimator.predict(X_test_scaled)

# Calculate R² score
r2 = r2_score(y_test, y_pred)
print(f"R² score: {r2}")
```

```python
import gradio as gr
import pandas as pd
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def predict_house_price(bed, bath, acre_lot, house_size, year, zip_code):
    # Create a new sample house
    new_house = pd.DataFrame({
        'bed': [bed],
        'bath': [bath],
        'acre_lot': [acre_lot],
        'house_size': [house_size],
        'year': [year],
        f'zip_{zip_code}': [1],  # Zip code handling
    }, columns=X.columns)

    # Fill missing zip code dummies with 0
    for col in X.columns:
        if col.startswith('zip_') and col not in new_house.columns:
            new_house[col] = 0

    # Check for NaNs before scaling
    if new_house.isnull().values.any():
        print("NaNs detected in input, filling with 0.")
        new_house.fillna(0, inplace=True)  # Fill with 0 or appropriate value

    # Ensure all columns are present
    new_house = new_house.reindex(columns=X.columns, fill_value=0)

    # Scale the new input data
    new_house_scaled = X_scaler.transform(new_house)

    # Predict the price
    predicted_price = estimator.predict(new_house_scaled)

    # **Fix for invalid index**: Assuming predicted_price is a 1D array
    return f"${predicted_price[0]:,.2f}"


# Create the Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("# House Price Prediction")

    bed = gr.Number(label="Bedrooms")
    bath = gr.Number(label="Bathrooms")
    acre_lot = gr.Number(label="Acre Lot")
    house_size = gr.Number(label="House Size (sq ft)")
    year = gr.Number(label="Year")
    zip_code = gr.Textbox(label="Zip Code")

    output = gr.Textbox(label="Predicted Price")

    predict_button = gr.Button("Predict Price")

    # When the button is clicked, make predictions
    predict_button.click(predict_house_price, inputs=[bed, bath, acre_lot, house_size, year, zip_code], outputs=output)

# Launch the Gradio interface
demo.launch(debug=True)
```
