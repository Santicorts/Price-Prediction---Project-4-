```python
import pandas as pd
import json 

# Path to the JSON file
json_file_path = 'Resources/2024_Listings.json'

# Load JSON file into a DataFrame
df = pd.read_json(json_file_path)

# Display the DataFrame
df
```

```python



# Load the JSON file
with open(json_file_path, 'r') as f:
    json_data = json.load(f)

# Print a sample of the loaded data to understand its structure
print(type(json_data))  # Check the type of the loaded data
print(json_data.keys())  # Print the top-level keys of the JSON data
print(json_data["0"])    # Print the first element to inspect the structure

```

```python
# Convert the dictionary values (listings) to a DataFrame
# Filter out keys that are not related to actual listings (e.g., remove 'dataType')
# We can do this by keeping only numeric keys
filtered_data = {k: v for k, v in json_data.items() if k.isdigit()}

# Convert the filtered dictionary (listings) to a DataFrame
df = pd.DataFrame.from_dict(filtered_data, orient='index')

pd.set_option('display.max_columns', None)

# Display the DataFrame to ensure the data is row-wise
print(df)
```
